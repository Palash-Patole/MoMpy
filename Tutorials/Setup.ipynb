# Setup

There are two ways to setup this toolbox on your system.

1. By installing the python packages manually
2. By setting up a conda environment with the provided environment file

Once the environment is created, it might be required to update it from time to time. This can be done by manually installing the newly required python package(s) in such a conda environment (or the global environment itself). However, an updated environment file can be easily used to update the conda environment with such new requirements. This is described [here](#update_conda_env). 


## By installing the python packages manually

Following the names of python packages that used in this toolbox. 

1. __[Numpy](https://numpy.org/install/)__
2. __[Matplotlib](https://matplotlib.org/3.1.1/users/installing.html)__
3. __[spiceypy](https://github.com/AndrewAnnex/SpiceyPy)__
4. __[pygmo](https://esa.github.io/pygmo2/install.html)__


It is possible to install all (or the missing) packages within the global environment using following commands.

| Python package | pip | conda|
| :- | :- | :- |
| numpy | pip install numpy | conda install numpy |
|matplotlib|pip install matplotlib| |
| spiceypy |pip install spiceypy|conda install spiceypy (from conda-forge)|
|pygmo|pip install pygmo|conda install pygmo (from conda-forge)|

## By setting up a conda environment with the provided environment file

One can follow the procedure given below to create a conda environment using the environment file MoMpy_env.yml, provided with this toolbox.

<div class="alert alert-block alert-info">
<b>Tip:</b> Similar information can be found at the official Anaconda documentation <a href='https://docs.conda.io/projects/conda/en/latest/user-guide/tasks/manage-environments.html#creating-an-environment-from-an-environment-yml-file'>here</a>. Information about how to use a conda environment with the Visual Studio Code can be found <a href='https://code.visualstudio.com/docs/python/environments#_conda-environments'>here </a>.
</div>

1. Open the Anaconda prompt.
2. Navigate to the working directory where MoMpy_env.yml file is present.
3. Run the following command: <br>
> conda env create -f MoMpy_env.yml
4. After successful creation of the environment, the commands to activate and deactivate the environment will be shown. Before activating the environment, it is possible to check the list of present environments using following command: <br>
> conda env list
5. Activate the environment with following command: <br>
> conda activate MoMpy_env
6. After this, the user can launch the IDE such as Spyder with the following command and edit/create new scripts: <br>
> spyder

  Or, the user can launch jupyter notebook, to use the notebooks provided in this (_MoMPy/Tutorials/_) folder using the following command: <br>
> jupyter notebook 
7. It is possible to deactivate the environment from the Anaconda prompt, with the following command:
> conda deactivate

### Updating existing conda environment 
<a id='update_conda_env'> </a>

1. Open the Anaconda prompt.
2. Navigate to the working directory where MoMpy_env.yml file is present.
3. Run the command:
>    conda env update --file MoMpy_env.yml --prune

   Note: The --prune option causes conda to remove any dependencies that are no longer required from the environment.
