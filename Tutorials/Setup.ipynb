# Setup

There are two ways to setup this toolbox on your system.

1. By installing the python packages manually
2. By setting up a conda environment with the provided environment file

## By installing the python packages manually

Following the names of python packages that used in this toolbox. 

1. Numpy
2. Matplotlib
3. spiceypy
4. PyGMO

| Python package | pip | conda|
| :- | :- | :- |
| numpy | pip install numpy | conda install numpy |

## By setting up a conda environment with the provided environment file

One can follow the procedure given below to create a conda environment using the environment file MoMpy_env.yml, provided with this toolbox.

1. Open the Anaconda prompt.
2. Navigate to the working directory where MoMpy_env.yml file is present.
3. Run the following command:
$$ \text{conda env create -f <MoMpy_env.yml>} $$
4. Check the list of present environments:
$$ \text{conda env list}$$
5. Activate the environment with following command
$$ \text{conda activate MoMpy_env}$$
